```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
data = pd.read_csv('titanic.csv')
data.isnull().sum()
data["Age"].fillna(data["Age"].mean(),inplace=True)
```

```python
sns.histplot(x="Age",data=data,bins=[i for i in range(0,81,10)])
mn = np.mean(data["Age"])
plt.plot([mn for i in range(0,300)],[i for i in range(0,300)],c ="red")
```

```python
print(np.median(data["Fare"]))
print(data["Fare"].median())
print(data["Fare"].mean())
sns.histplot(x="Fare",data=data,bins=[i for i in range(0,81,10)])
md = np.mean(data["Fare"])
mx = np.median(data["Fare"])
mo = data["Fare"].mode()[0]
plt.plot([md for i in range(0,300)],[i for i in range(0,300)],c ="green",label="mean")
plt.plot([mx for i in range(0,300)],[i for i in range(0,300)],c ="blue",label="mean")
plt.plot([mo for i in range(0,300)],[i for i in range(0,300)],c ="red",label="mean")
plt.legend()
```
